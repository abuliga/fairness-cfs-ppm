```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch
no_fairness = pd.read_csv('/Users/andrei/Desktop/FBK/fairness_cfs/top_k_results/cf_analysis_results/kappel_logs_counterfactual_results_without_fairness_optimization_diff_models_simple_trace_update_constraints.csv')
fairness_with_plaus = pd.read_csv('/Users/andrei/Desktop/FBK/fairness_cfs/top_k_results/cf_analysis_results/kappel_logs_counterfactual_results_with_fairness_optimization_diff_models_simple_trace_update_constraints.csv')

# Add optimization column
no_fairness['optimization'] = 'No Fairness'
fairness_with_plaus['optimization'] = 'Fairness'

# Combine the two DataFrames
combined = pd.concat([no_fairness, fairness_with_plaus])

# Create combined hue: "model | case"
combined['model_case'] = combined['model'] + ' | ' + combined['case']

# Metrics and layout
metrics = ['proximity', 'alternative_sparsity', 'conformance', 'plausibility', 'unfairness', 'top_k_diversity']
datasets = combined['dataset'].unique()
models = sorted(combined['model'].unique())

# Hue order: for each model, protected first then unprotected
hue_order = []
for model in models:
    hue_order.append(f'{model} | protected')
    hue_order.append(f'{model} | unprotected')

# Color palette: same color per model, lighter shade for unprotected
base_palette = sns.color_palette('tab10', len(models))
model_color_map = {model: color for model, color in zip(models, base_palette)}

def lighten_color(color, amount=0.45):
    import colorsys
    c = colorsys.rgb_to_hls(*color)
    return colorsys.hls_to_rgb(c[0], min(1, c[1] + amount), c[2])

palette = {}
for model, color in model_color_map.items():
    palette[f'{model} | protected'] = color
    palette[f'{model} | unprotected'] = lighten_color(color)

# Plot
fig, axes = plt.subplots(nrows=len(datasets[:2]), ncols=len(metrics), figsize=(30, 4 * len(datasets[:2])))

for i, dataset in enumerate(datasets[:2]):
    for j, metric in enumerate(metrics):
        ax = axes[i, j]
        data = combined[combined['dataset'] == dataset]

        sns.boxplot(
            data=data,
            x='optimization',
            y=metric,
            hue='model_case',
            hue_order=hue_order,
            palette=palette,
            order=['No Fairness', 'Fairness'],
            ax=ax,
            linewidth=0.8,
        )

        ax.set_title(metric, fontsize=15)
        ax.set_xlabel('')
        ax.set_ylabel('')
        if j == 0:
            ax.set_ylabel(dataset, fontsize=14, fontweight='bold')
        ax.get_legend().remove()
        ax.tick_params(axis='x', labelsize=9)

# Legend: dark shade = protected, light shade = unprotected, color = model
legend_elements = []
for model, color in model_color_map.items():
    legend_elements.append(Patch(facecolor=color, edgecolor='grey', label=f'{model} | protected'))
    legend_elements.append(Patch(facecolor=lighten_color(color), edgecolor='grey', label=f'{model} | unprotected'))

fig.legend(
    handles=legend_elements,
    title='Model | Case',
    bbox_to_anchor=(0.5, 1.01),
    loc='lower center',
    ncol=len(models) * 2,
    fontsize=16,
    title_fontsize=13,
    frameon=True,
)

plt.subplots_adjust(top=0.93, bottom=0.05, left=0.05, right=0.98, hspace=0.4, wspace=0.35)
plt.show()
```

```python
cols = ['proximity','conformance','alternative_sparsity','plausibility','unfairness','top_k_diversity']

def compute_gap(df):
    g = df[df['model']=='mlp'].groupby(['dataset','case'])[cols].mean().unstack()

    gap = g.xs('protected', level=1, axis=1) - g.xs('unprotected', level=1, axis=1)
    return gap.reset_index()
```

```python
cols = ['proximity','conformance','alternative_sparsity','plausibility','unfairness']

def prepare(df, method):
    res = (
        df[(df['model']=='mlp') & (df['case']=='protected')]
        .groupby('dataset')[cols]
        .mean()
        .reset_index()
    )
    res['method'] = method
    return res

data = pd.concat([
    prepare(fairness, 'fairness'),
    prepare(no_fairness, 'no_fairness'),
    prepare(fixed_sensitive_attributes, 'fixed_sensitive')
])
```

```python
data.rename(columns={'alternative_sparsity':'sparsity'},inplace=True)

ranked = data[data['dataset']!='hb_-age_-gender'].copy()
cols = ['proximity','conformance','sparsity','plausibility','unfairness']

ranked['proximity_rank'] = ranked.groupby('dataset')['proximity'].rank(ascending=True)
ranked['sparsity_rank'] = ranked.groupby('dataset')['sparsity'].rank(ascending=True)
ranked['plausibility_rank'] = ranked.groupby('dataset')['plausibility'].rank(ascending=True)

ranked['conformance_rank'] = ranked.groupby('dataset')['conformance'].rank(ascending=False)

ranked['unfairness_rank'] = ranked.groupby('dataset')['unfairness'].rank(ascending=True)
```

```python
table = ranked[['dataset','method']].copy()

for m in cols:
    table[m] = (
        ranked[m].round(3).astype(str) +
        " (" + ranked[m+'_rank'].astype(int).astype(str) + ")"
    )
```

```python
final_table = table.pivot(index='dataset', columns='method', values=cols)

final_table
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.gridspec import GridSpec
import ast

# ── File paths ──────────────────────────────────────────────────────────────
MODEL_RESULTS_FILE  = "model_results.csv"
MLP_FLIP_FILE       = "synthetic_logs/protected_label/mlp_fairness_flip_results_hospital_log_high-xes_simple_trace_fixed_sensitive_attrs.csv"
XGB_FLIP_FILE       = "synthetic_logs/protected_label/xgboost_fairness_flip_results_hospital_log_high-xes_simple_trace_fixed_sensitive_attrs.csv"

# ── Colour palette ───────────────────────────────────────────────────────────
COLORS = {
    ("mlp",     False): "#5B8DB8",   # blue-ish, light
    ("mlp",     True):  "#1A4E78",   # blue-ish, dark
    ("xgboost", False): "#E8945A",   # orange, light
    ("xgboost", True):  "#9E4B0E",   # orange, dark
}
PROTECTED_COLOR   = "#C0392B"
UNPROTECTED_COLOR = "#2980B9"
TOTAL_COLOR       = "#27AE60"


# ════════════════════════════════════════════════════════════════════════════
# 1. Load & reshape model_results.csv
# ════════════════════════════════════════════════════════════════════════════
raw = pd.read_csv(MODEL_RESULTS_FILE, header=None)

# Row 0 → fairness_opt values (cols 3-6)
# Row 1 → model names       (cols 3-6)
# Rows 2+ → data
fairness_opts = [False, False, True, True]
models        = ["mlp", "xgboost", "mlp", "xgboost"]

data_rows = raw.iloc[3:].copy()
data_rows.columns = ["case", "prefix_length", "changed_attributes",
                     "val_mlp_False", "val_xgboost_False",
                     "val_mlp_True",  "val_xgboost_True"]
data_rows["prefix_length"] = data_rows["prefix_length"].astype(int)
for col in ["val_mlp_False", "val_xgboost_False", "val_mlp_True", "val_xgboost_True"]:
    data_rows[col] = pd.to_numeric(data_rows[col], errors="coerce").fillna(0.0)

# Parse changed_attributes into a frozen-set for grouping
def parse_attrs(s):
    try:
        lst = ast.literal_eval(str(s))
        return frozenset(lst)
    except Exception:
        return frozenset()

data_rows["attr_set"] = data_rows["changed_attributes"].apply(parse_attrs)

# Count how many attributes were changed (0 = no-change counterfactual)
data_rows["n_changed"] = data_rows["attr_set"].apply(len)


# ════════════════════════════════════════════════════════════════════════════
# 2. FIGURE 1 – Effect of fairness_opt on # changed attributes
#    Grouped bar chart: for each prefix length, show mean % share of
#    counterfactuals that changed exactly 1, 2, 3+ attributes
#    split by model × fairness_opt.
# ════════════════════════════════════════════════════════════════════════════
def aggregate_by_n_changed(df, val_col):
    """Return df with prefix_length × n_changed → share (%)."""
    total_per_prefix = df.groupby("prefix_length")[val_col].sum()
    grouped = df.groupby(["prefix_length", "n_changed"])[val_col].sum()
    share   = (grouped / total_per_prefix * 100).reset_index()
    share.columns = ["prefix_length", "n_changed", "share"]
    return share


configs = [
    ("mlp",     False, "val_mlp_False"),
    ("mlp",     True,  "val_mlp_True"),
    ("xgboost", False, "val_xgboost_False"),
    ("xgboost", True,  "val_xgboost_True"),
]

prefix_lengths = sorted(data_rows["prefix_length"].unique())
n_groups       = len(prefix_lengths)


# ════════════════════════════════════════════════════════════════════════════
# 3. FIGURE 2 – Single-attribute dominance: % share of ['age']-only CFs
#    across prefix lengths, fairness_opt × model
# ════════════════════════════════════════════════════════════════════════════
age_only = data_rows[data_rows["attr_set"] == frozenset(["age"])].copy()
all_pl   = data_rows.groupby("prefix_length")[["val_mlp_False","val_xgboost_False",
                                                "val_mlp_True", "val_xgboost_True"]].sum()
age_pl   = age_only.groupby("prefix_length")[["val_mlp_False","val_xgboost_False",
                                               "val_mlp_True", "val_xgboost_True"]].sum()

age_share = (age_pl / all_pl * 100).reindex(prefix_lengths).fillna(0)

fig2, ax = plt.subplots(figsize=(9, 4))
ax.set_title("Share of counterfactuals changing only 'age'\nacross prefix lengths, by model × fairness_opt",
             fontsize=12, fontweight="bold")

for model, fairness_opt, col in configs:
    ax.plot(
        prefix_lengths,
        age_share[col].values,
        marker="o",
        linewidth=2,
        color=COLORS[(model, fairness_opt)],
        linestyle="-" if not fairness_opt else "--",
        label=f"{model.upper()}, opt={fairness_opt}",
    )

ax.set_xlabel("Prefix length", fontsize=10)
ax.set_ylabel("% of all CFs", fontsize=10)
ax.set_xticks(prefix_lengths)
ax.yaxis.grid(True, linestyle="--", alpha=0.4)
ax.set_axisbelow(True)
ax.legend(fontsize=9)
plt.tight_layout()
plt.savefig("fig2_age_only_share.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved fig2_age_only_share.png")


# ════════════════════════════════════════════════════════════════════════════
# 4. Load flip results (outcome flip using specific sensitive attributes)
# ════════════════════════════════════════════════════════════════════════════
mlp_flip = pd.read_csv(MLP_FLIP_FILE)
xgb_flip = pd.read_csv(XGB_FLIP_FILE)

# The XGB file has repeated blocks (same attribute combos appear multiple times)
# – deduplicate by keeping unique attribute rows with a non-zero total rate
# (rows with all zeros = MLP-style, rows with values = XGB proper results)
xgb_nonzero = xgb_flip[xgb_flip["cf_rate_total"] > 0].drop_duplicates(subset="attribute")


# ════════════════════════════════════════════════════════════════════════════
# 5. FIGURE 3 – Protected flip rate: single attributes, MLP vs XGBoost
# ════════════════════════════════════════════════════════════════════════════
mlp_nonzero = mlp_flip[mlp_flip["cf_rate_total"] > 0].drop_duplicates(subset="attribute")

mlp_single = mlp_nonzero[~mlp_nonzero["attribute"].str.contains(r"\+")].copy()
xgb_single = xgb_nonzero[~xgb_nonzero["attribute"].str.contains(r"\+")].copy()

# Align on the union of attribute sets
all_single_attrs = sorted(set(mlp_single["attribute"]) | set(xgb_single["attribute"]))
mlp_s = mlp_single.set_index("attribute").reindex(all_single_attrs).fillna(0)
xgb_s = xgb_single.set_index("attribute").reindex(all_single_attrs).fillna(0)

# Sort by XGBoost protected rate
order3 = xgb_s["cf_rate_protected"].sort_values(ascending=True).index
mlp_s  = mlp_s.loc[order3]
xgb_s  = xgb_s.loc[order3]

y3 = np.arange(len(order3))
h3 = 0.35

fig3, ax = plt.subplots(figsize=(9, 5))
ax.set_title("Outcome flip rate – protected group, single sensitive attributes\nMLP vs XGBoost",
             fontsize=12, fontweight="bold")

ax.barh(y3 + h3/2, xgb_s["cf_rate_protected"].values, h3,
        color=COLORS[("xgboost", False)], alpha=0.85, label="XGBoost")
ax.barh(y3 - h3/2, mlp_s["cf_rate_protected"].values, h3,
        color=COLORS[("mlp", False)],     alpha=0.85, label="MLP")

ax.set_yticks(y3)
ax.set_yticklabels(order3.tolist(), fontsize=10)
ax.set_xlabel("CF flip rate (protected)", fontsize=10)
ax.xaxis.grid(True, linestyle="--", alpha=0.4)
ax.set_axisbelow(True)
ax.legend(fontsize=9)
plt.tight_layout()
plt.savefig("fig3_flip_single_attr_both_models.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved fig3_flip_single_attr_both_models.png")


# ════════════════════════════════════════════════════════════════════════════
# 6. FIGURE 4 – Protected flip rate: all multi-attribute combos WITHOUT age,
#    MLP vs XGBoost side by side, coloured by combo size
# ════════════════════════════════════════════════════════════════════════════
def get_multi_no_age(flip_df):
    """Multi-attribute combos ('+' present) that do NOT contain 'age'."""
    nonzero = flip_df[flip_df["cf_rate_total"] > 0].drop_duplicates(subset="attribute")
    multi   = nonzero[nonzero["attribute"].str.contains(r"\+")].copy()
    no_age  = multi[~multi["attribute"].str.contains(r"\bage\b", regex=True)].copy()
    no_age["n_attrs"] = no_age["attribute"].str.count(r"\+") + 2
    return no_age

mlp_multi = get_multi_no_age(mlp_flip)
xgb_multi = get_multi_no_age(xgb_flip)

# Align on union of combos
all_combos = sorted(set(mlp_multi["attribute"]) | set(xgb_multi["attribute"]))
mlp_m = mlp_multi.set_index("attribute").reindex(all_combos).fillna(0)
xgb_m = xgb_multi.set_index("attribute").reindex(all_combos).fillna(0)

# Sort by XGBoost protected rate
order4 = xgb_m["cf_rate_protected"].sort_values(ascending=True).index
mlp_m  = mlp_m.loc[order4]
xgb_m  = xgb_m.loc[order4]

# Colour by number of attributes (use XGB metadata; default 2 for unknowns)
n_attrs_map = (
    xgb_multi.set_index("attribute")["n_attrs"]
    .reindex(order4).fillna(2).astype(int)
)
cmap_n = {2: "#5B8DB8", 3: "#E8945A", 4: "#8E44AD"}
bar_colors = [cmap_n.get(n, "#888") for n in n_attrs_map]

y4   = np.arange(len(order4))
h4   = 0.35
fig_h = max(6, len(order4) * 0.45)

fig4, ax = plt.subplots(figsize=(11, fig_h))
ax.set_title("Outcome flip rate – protected group, multi-attribute combos (no 'age')\nMLP vs XGBoost",
             fontsize=12, fontweight="bold")

ax.barh(y4 + h4/2, xgb_m["cf_rate_protected"].values, h4,
        color=bar_colors, alpha=0.85, label="XGBoost")
ax.barh(y4 - h4/2, mlp_m["cf_rate_protected"].values, h4,
        color=bar_colors, alpha=0.40, label="MLP")

ax.set_yticks(y4)
ax.set_yticklabels(order4.tolist(), fontsize=8)
ax.set_xlabel("CF flip rate (protected)", fontsize=10)
ax.xaxis.grid(True, linestyle="--", alpha=0.4)
ax.set_axisbelow(True)

legend_patches  = [mpatches.Patch(color=c, label=f"{n} attrs") for n, c in cmap_n.items()]
legend_patches += [mpatches.Patch(color="#aaa", alpha=0.85, label="XGBoost"),
                   mpatches.Patch(color="#aaa", alpha=0.40, label="MLP")]
ax.legend(handles=legend_patches, fontsize=9, loc="lower right")
plt.tight_layout()
plt.savefig("fig4_flip_multi_no_age_both_models.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved fig4_flip_multi_no_age_both_models.png")


# ════════════════════════════════════════════════════════════════════════════
# 7. FIGURE 5 – MLP vs XGBoost: fairness_opt effect summary heatmap
#    Rows = prefix lengths, Cols = models × fairness_opt
#    Cell = % share of single-attribute-only counterfactuals
# ════════════════════════════════════════════════════════════════════════════
single_only = data_rows[data_rows["n_changed"] == 1].copy()
total_per   = data_rows.groupby("prefix_length")[["val_mlp_False","val_xgboost_False",
                                                   "val_mlp_True","val_xgboost_True"]].sum()
single_per  = single_only.groupby("prefix_length")[["val_mlp_False","val_xgboost_False",
                                                     "val_mlp_True","val_xgboost_True"]].sum()
heatmap_df  = (single_per / total_per * 100).reindex(prefix_lengths).fillna(0).T

heatmap_df.index = ["MLP / opt=False", "XGB / opt=False",
                     "MLP / opt=True",  "XGB / opt=True"]
heatmap_df.columns = [f"PL={pl}" for pl in prefix_lengths]

fig5, ax = plt.subplots(figsize=(10, 3.5))
ax.set_title("% of counterfactuals changing exactly 1 attribute\n(single-attribute focus; rows = model×opt, cols = prefix length)",
             fontsize=12, fontweight="bold")

im = ax.imshow(heatmap_df.values, aspect="auto", cmap="YlOrRd", vmin=0, vmax=100)
plt.colorbar(im, ax=ax, label="% share")

ax.set_xticks(range(len(heatmap_df.columns)))
ax.set_xticklabels(heatmap_df.columns, fontsize=10)
ax.set_yticks(range(len(heatmap_df.index)))
ax.set_yticklabels(heatmap_df.index, fontsize=10)

for i in range(len(heatmap_df.index)):
    for j in range(len(heatmap_df.columns)):
        val = heatmap_df.values[i, j]
        ax.text(j, i, f"{val:.1f}%", ha="center", va="center",
                fontsize=9, color="black" if val < 60 else "white")

plt.tight_layout()
plt.savefig("fig5_heatmap_single_attr_share.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved fig5_heatmap_single_attr_share.png")
```

```python
ORS = {
    ("mlp",     False): "#5B8DB8",   # blue-ish, light
    ("mlp",     True):  "#1A4E78",   # blue-ish, dark
    ("xgboost", False): "#E8945A",   # orange, light
    ("xgboost", True):  "#9E4B0E",   # orange, dark
}
PROTECTED_COLOR   = "#C0392B"
UNPROTECTED_COLOR = "#2980B9"
TOTAL_COLOR       = "#27AE60"


# ════════════════════════════════════════════════════════════════════════════
# 1. Load & reshape model_results.csv
# ════════════════════════════════════════════════════════════════════════════
raw = pd.read_csv(MODEL_RESULTS_FILE, header=None)

# Row 0 → fairness_opt values (cols 3-6)
# Row 1 → model names       (cols 3-6)
# Rows 2+ → data
fairness_opts = [False, False, True, True]
models        = ["mlp", "xgboost", "mlp", "xgboost"]

data_rows = raw.iloc[3:].copy()
data_rows.columns = ["case", "prefix_length", "changed_attributes",
                     "val_mlp_False", "val_xgboost_False",
                     "val_mlp_True",  "val_xgboost_True"]
data_rows["prefix_length"] = data_rows["prefix_length"].astype(int)
for col in ["val_mlp_False", "val_xgboost_False", "val_mlp_True", "val_xgboost_True"]:
    data_rows[col] = pd.to_numeric(data_rows[col], errors="coerce").fillna(0.0)

# Parse changed_attributes into a frozen-set for grouping
def parse_attrs(s):
    try:
        lst = ast.literal_eval(str(s))
        return frozenset(lst)
    except Exception:
        return frozenset()

data_rows["attr_set"] = data_rows["changed_attributes"].apply(parse_attrs)

# Count how many attributes were changed (0 = no-change counterfactual)
data_rows["n_changed"] = data_rows["attr_set"].apply(len)


# ════════════════════════════════════════════════════════════════════════════
# 2. FIGURE 1 – Effect of fairness_opt on # changed attributes
#    Grouped bar chart: for each prefix length, show mean % share of
#    counterfactuals that changed exactly 1, 2, 3+ attributes
#    split by model × fairness_opt.
# ════════════════════════════════════════════════════════════════════════════
def aggregate_by_n_changed(df, val_col):
    """Return df with prefix_length × n_changed → share (%)."""
    total_per_prefix = df.groupby("prefix_length")[val_col].sum()
    grouped = df.groupby(["prefix_length", "n_changed"])[val_col].sum()
    share   = (grouped / total_per_prefix * 100).reset_index()
    share.columns = ["prefix_length", "n_changed", "share"]
    return share


configs = [
    ("mlp",     False, "val_mlp_False"),
    ("mlp",     True,  "val_mlp_True"),
    ("xgboost", False, "val_xgboost_False"),
    ("xgboost", True,  "val_xgboost_True"),
]

prefix_lengths = sorted(data_rows["prefix_length"].unique())
n_groups       = len(prefix_lengths)

fig1, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
fig1.suptitle("Effect of fairness_opt on number of changed attributes\n(% of counterfactuals per prefix length)",
              fontsize=13, fontweight="bold", y=1.02)

for ax_idx, model in enumerate(["mlp", "xgboost"]):
    ax = axes[ax_idx]
    ax.set_title(f"Model: {model.upper()}", fontsize=11, fontweight="bold")

    bar_width = 0.35
    x = np.arange(n_groups)

    for opt_idx, fairness_opt in enumerate([False, True]):
        val_col   = f"val_{model}_{'True' if fairness_opt else 'False'}"
        agg       = aggregate_by_n_changed(data_rows, val_col)

        # Stack bars: 0-change (grey), 1, 2, 3+
        bottoms = np.zeros(n_groups)
        n_change_cats = [0, 1, 2, 3]
        alphas        = [0.3, 0.7, 0.85, 1.0]
        labels_nc     = ["0 attrs (no change)", "1 attr", "2 attrs", "3+ attrs"]

        for nc, alpha, lbl in zip(n_change_cats, alphas, labels_nc):
            shares = []
            for pl in prefix_lengths:
                row = agg[(agg["prefix_length"] == pl) & (agg["n_changed"] == nc)]
                shares.append(row["share"].values[0] if len(row) else 0.0)
            shares = np.array(shares)
            color  = COLORS[(model, fairness_opt)]

            bar = ax.bar(
                x + opt_idx * bar_width,
                shares,
                bar_width,
                bottom=bottoms,
                color=color,
                alpha=alpha,
                label=f"opt={fairness_opt}, {lbl}" if ax_idx == 0 else None,
            )
            bottoms += shares

    ax.set_xticks(x + bar_width / 2)
    ax.set_xticklabels([f"PL={pl}" for pl in prefix_lengths], fontsize=9)
    ax.set_xlabel("Prefix length", fontsize=10)
    if ax_idx == 0:
        ax.set_ylabel("Share of counterfactuals (%)", fontsize=10)
    ax.set_ylim(0, 110)
    ax.yaxis.grid(True, linestyle="--", alpha=0.4)
    ax.set_axisbelow(True)

    # Mini-legend for False / True
    p1 = mpatches.Patch(color=COLORS[(model, False)], label="fairness_opt=False")
    p2 = mpatches.Patch(color=COLORS[(model, True)],  label="fairness_opt=True")
    ax.legend(handles=[p1, p2], fontsize=9, loc="upper right")

plt.tight_layout()
plt.savefig("fig1_n_changed_attrs.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved fig1_n_changed_attrs.png")


# ════════════════════════════════════════════════════════════════════════════
# 3. FIGURE 2 – Single-attribute dominance: % share of ['age']-only CFs
#    across prefix lengths, fairness_opt × model
# ════════════════════════════════════════════════════════════════════════════
age_only = data_rows[data_rows["attr_set"] == frozenset(["age"])].copy()
all_pl   = data_rows.groupby("prefix_length")[["val_mlp_False","val_xgboost_False",
                                                "val_mlp_True", "val_xgboost_True"]].sum()
age_pl   = age_only.groupby("prefix_length")[["val_mlp_False","val_xgboost_False",
                                               "val_mlp_True", "val_xgboost_True"]].sum()

age_share = (age_pl / all_pl * 100).reindex(prefix_lengths).fillna(0)

fig2, ax = plt.subplots(figsize=(9, 4))
ax.set_title("Share of counterfactuals changing only 'age'\nacross prefix lengths, by model × fairness_opt",
             fontsize=12, fontweight="bold")

for model, fairness_opt, col in configs:
    ax.plot(
        prefix_lengths,
        age_share[col].values,
        marker="o",
        linewidth=2,
        color=COLORS[(model, fairness_opt)],
        linestyle="-" if not fairness_opt else "--",
        label=f"{model.upper()}, opt={fairness_opt}",
    )

ax.set_xlabel("Prefix length", fontsize=10)
ax.set_ylabel("% of all CFs", fontsize=10)
ax.set_xticks(prefix_lengths)
ax.yaxis.grid(True, linestyle="--", alpha=0.4)
ax.set_axisbelow(True)
ax.legend(fontsize=9)
plt.tight_layout()
plt.savefig("fig2_age_only_share.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved fig2_age_only_share.png")


# ════════════════════════════════════════════════════════════════════════════
# 4. Load flip results (outcome flip using specific sensitive attributes)
# ════════════════════════════════════════════════════════════════════════════
mlp_flip = pd.read_csv(MLP_FLIP_FILE)
xgb_flip = pd.read_csv(XGB_FLIP_FILE)

# The XGB file has repeated blocks (same attribute combos appear multiple times)
# – deduplicate by keeping unique attribute rows with a non-zero total rate
# (rows with all zeros = MLP-style, rows with values = XGB proper results)
xgb_nonzero = xgb_flip[xgb_flip["cf_rate_total"] > 0].drop_duplicates(subset="attribute")


# ════════════════════════════════════════════════════════════════════════════
# 5. FIGURE 3 – Protected flip rate: single attributes, MLP vs XGBoost
# ════════════════════════════════════════════════════════════════════════════
mlp_nonzero = mlp_flip[mlp_flip["cf_rate_total"] > 0].drop_duplicates(subset="attribute")

mlp_single = mlp_nonzero[~mlp_nonzero["attribute"].str.contains(r"\+")].copy()
xgb_single = xgb_nonzero[~xgb_nonzero["attribute"].str.contains(r"\+")].copy()

# Align on the union of attribute sets
all_single_attrs = sorted(set(mlp_single["attribute"]) | set(xgb_single["attribute"]))
mlp_s = mlp_single.set_index("attribute").reindex(all_single_attrs).fillna(0)
xgb_s = xgb_single.set_index("attribute").reindex(all_single_attrs).fillna(0)

# Sort by XGBoost protected rate
order3 = xgb_s["cf_rate_protected"].sort_values(ascending=True).index
mlp_s  = mlp_s.loc[order3]
xgb_s  = xgb_s.loc[order3]

y3 = np.arange(len(order3))
h3 = 0.35

fig3, ax = plt.subplots(figsize=(9, 5))
ax.set_title("Outcome flip rate – protected group, single sensitive attributes\nMLP vs XGBoost",
             fontsize=12, fontweight="bold")

ax.barh(y3 + h3/2, xgb_s["cf_rate_protected"].values, h3,
        color=COLORS[("xgboost", False)], alpha=0.85, label="XGBoost")
ax.barh(y3 - h3/2, mlp_s["cf_rate_protected"].values, h3,
        color=COLORS[("mlp", False)],     alpha=0.85, label="MLP")

ax.set_yticks(y3)
ax.set_yticklabels(order3.tolist(), fontsize=10)
ax.set_xlabel("CF flip rate (protected)", fontsize=10)
ax.xaxis.grid(True, linestyle="--", alpha=0.4)
ax.set_axisbelow(True)
ax.legend(fontsize=9)
plt.tight_layout()
plt.savefig("fig3_flip_single_attr_both_models.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved fig3_flip_single_attr_both_models.png")


# ════════════════════════════════════════════════════════════════════════════
# 6. FIGURE 4 – Protected flip rate: all multi-attribute combos WITHOUT age,
#    MLP vs XGBoost side by side — same colours as Fig 3 (MLP blue, XGB orange)
# ════════════════════════════════════════════════════════════════════════════
def get_multi_no_age(flip_df):
    """Multi-attribute combos ('+' present) that do NOT contain 'age'."""
    nonzero = flip_df[flip_df["cf_rate_total"] > 0].drop_duplicates(subset="attribute")
    multi   = nonzero[nonzero["attribute"].str.contains(r"\+")].copy()
    no_age  = multi[~multi["attribute"].str.contains(r"\bage\b", regex=True)].copy()
    return no_age

mlp_multi = get_multi_no_age(mlp_flip)
xgb_multi = get_multi_no_age(xgb_flip)

# Align on union of combos
all_combos = sorted(set(mlp_multi["attribute"]) | set(xgb_multi["attribute"]))
mlp_m = mlp_multi.set_index("attribute").reindex(all_combos).fillna(0)
xgb_m = xgb_multi.set_index("attribute").reindex(all_combos).fillna(0)

# Sort by XGBoost protected rate
order4 = xgb_m["cf_rate_protected"].sort_values(ascending=True).index
mlp_m  = mlp_m.loc[order4]
xgb_m  = xgb_m.loc[order4]

y4    = np.arange(len(order4))
h4    = 0.35
fig_h = max(6, len(order4) * 0.45)

fig4, ax = plt.subplots(figsize=(11, fig_h))
ax.set_title("Outcome flip rate – protected group, multi-attribute combos (no 'age')\nMLP vs XGBoost",
             fontsize=12, fontweight="bold")

ax.barh(y4 + h4/2, xgb_m["cf_rate_protected"].values, h4,
        color=COLORS[("xgboost", False)], alpha=0.85, label="XGBoost")
ax.barh(y4 - h4/2, mlp_m["cf_rate_protected"].values, h4,
        color=COLORS[("mlp", False)],     alpha=0.85, label="MLP")

ax.set_yticks(y4)
ax.set_yticklabels(order4.tolist(), fontsize=8)
ax.set_xlabel("CF flip rate (protected)", fontsize=10)
ax.xaxis.grid(True, linestyle="--", alpha=0.4)
ax.set_axisbelow(True)
ax.legend(fontsize=9)
plt.tight_layout()
plt.savefig("fig4_flip_multi_no_age_both_models.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved fig4_flip_multi_no_age_both_models.png")


# ════════════════════════════════════════════════════════════════════════════
# 8. FIGURE 6 – Per-attribute % share of CFs by prefix length
#    2 rows (MLP / XGBoost) × 2 cols (fairness_opt False / True)
#    Grouped bars: one bar-group per prefix length, one bar per attribute
# ════════════════════════════════════════════════════════════════════════════

# Collect all individual attributes that appear in changed_attributes
all_attrs_in_data = set()
for s in data_rows["attr_set"]:
    all_attrs_in_data.update(s)
all_attrs_in_data = sorted(all_attrs_in_data)

# For each (model, fairness_opt, prefix_length, attribute) compute
# what % of CFs include that attribute (single-attr rows only)
ATTR_COLORS = {
    "age":                  "#E24B4A",
    "citizen":              "#3266AD",
    "gender":               "#27AE60",
    "german speaking":      "#E8945A",
    "private_insurance":    "#8E44AD",
    "underlying_condition": "#16A085",
}
default_color = "#888888"

fig6, axes6 = plt.subplots(2, 2, figsize=(16, 10), sharey=False)
fig6.suptitle("% share of CFs per changed attribute, by prefix length\n(single-attribute counterfactuals only)",
              fontsize=13, fontweight="bold", y=1.01)

x6        = np.arange(len(prefix_lengths))
n_attrs6  = len(all_attrs_in_data)
bar_w6    = 0.8 / n_attrs6          # total bar group width = 0.8

for row_idx, model in enumerate(["mlp", "xgboost"]):
    for col_idx, fairness_opt in enumerate([False, True]):
        ax  = axes6[row_idx][col_idx]
        val_col = f"val_{model}_{'True' if fairness_opt else 'False'}"

        # Total CFs per prefix length (for denominator)
        total_per_pl = data_rows.groupby("prefix_length")[val_col].sum()

        ax.set_title(f"{model.upper()} – fairness_opt={fairness_opt}",
                     fontsize=11, fontweight="bold")

        for a_idx, attr in enumerate(all_attrs_in_data):
            attr_rows = data_rows[data_rows["attr_set"] == frozenset([attr])]
            attr_per_pl = attr_rows.groupby("prefix_length")[val_col].sum().reindex(prefix_lengths).fillna(0)
            shares = (attr_per_pl.values / total_per_pl.reindex(prefix_lengths).values * 100)

            offset = (a_idx - n_attrs6 / 2 + 0.5) * bar_w6
            ax.bar(x6 + offset, shares, bar_w6,
                   color=ATTR_COLORS.get(attr, default_color),
                   alpha=0.85,
                   label=attr if (row_idx == 0 and col_idx == 0) else None)

        ax.set_xticks(x6)
        ax.set_xticklabels([f"PL={pl}" for pl in prefix_lengths], fontsize=9)
        ax.set_xlabel("Prefix length", fontsize=10)
        ax.set_ylabel("% of all CFs", fontsize=10)
        ax.yaxis.grid(True, linestyle="--", alpha=0.4)
        ax.set_axisbelow(True)

# Shared legend from first subplot
handles, labels = axes6[0][0].get_legend_handles_labels()
fig6.legend(handles, labels, loc="lower center", ncol=len(all_attrs_in_data),
            fontsize=9, bbox_to_anchor=(0.5, -0.04), title="Attribute")

plt.tight_layout()
plt.savefig("fig6_attr_share_by_prefix_fairness.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved fig6_attr_share_by_prefix_fairness.png")


# ════════════════════════════════════════════════════════════════════════════
# 7. FIGURE 5 – MLP vs XGBoost: fairness_opt effect summary heatmap
#    Rows = prefix lengths, Cols = models × fairness_opt
#    Cell = % share of single-attribute-only counterfactuals
# ════════════════════════════════════════════════════════════════════════════
single_only = data_rows[data_rows["n_changed"] == 1].copy()
total_per   = data_rows.groupby("prefix_length")[["val_mlp_False","val_xgboost_False",
                                                   "val_mlp_True","val_xgboost_True"]].sum()
single_per  = single_only.groupby("prefix_length")[["val_mlp_False","val_xgboost_False",
                                                     "val_mlp_True","val_xgboost_True"]].sum()
heatmap_df  = (single_per / total_per * 100).reindex(prefix_lengths).fillna(0).T

heatmap_df.index = ["MLP / opt=False", "XGB / opt=False",
                     "MLP / opt=True",  "XGB / opt=True"]
heatmap_df.columns = [f"PL={pl}" for pl in prefix_lengths]

fig5, ax = plt.subplots(figsize=(10, 3.5))
ax.set_title("% of counterfactuals changing exactly 1 attribute\n(single-attribute focus; rows = model×opt, cols = prefix length)",
             fontsize=12, fontweight="bold")

im = ax.imshow(heatmap_df.values, aspect="auto", cmap="YlOrRd", vmin=0, vmax=100)
plt.colorbar(im, ax=ax, label="% share")

ax.set_xticks(range(len(heatmap_df.columns)))
ax.set_xticklabels(heatmap_df.columns, fontsize=10)
ax.set_yticks(range(len(heatmap_df.index)))
ax.set_yticklabels(heatmap_df.index, fontsize=10)

for i in range(len(heatmap_df.index)):
    for j in range(len(heatmap_df.columns)):
        val = heatmap_df.values[i, j]
        ax.text(j, i, f"{val:.1f}%", ha="center", va="center",
                fontsize=9, color="black" if val < 60 else "white")

plt.tight_layout()
plt.savefig("fig5_heatmap_single_attr_share.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved fig5_heatmap_single_attr_share.png")
```

```python
import pandas as pd
import numpy as np

# =========================
# 1. LOAD DATA
# =========================

prot_df = pd.read_csv("categorize_attributes_final/bpi2012_test_cases_protected.csv")
unprot_df = pd.read_csv("categorize_attributes_final/bpi2012_test_cases_unprotected.csv")


# =========================
# 2. PREPARE ORIGINAL DATA
# =========================
prot_df['trace_id'] = prot_df.index
prot_df['group'] = 'protected'

unprot_df['trace_id'] = unprot_df.index
unprot_df['group'] = 'unprotected'

orig_df = pd.concat([prot_df, unprot_df], ignore_index=True)


# =========================
# 3. MERGE CF + ORIGINAL
# =========================
merged = full_cf_df.merge(
    orig_df[['trace_id', 'AMOUNT_REQ', 'group']],
    on='trace_id',
    how='left'
)

# after merge:
# AMOUNT_REQ_x = CF
# AMOUNT_REQ_y = original

# safety rename (cleaner)
merged = merged.rename(columns={
    'AMOUNT_REQ_x': 'amount_cf',
    'AMOUNT_REQ_y': 'amount_orig'
})

assert merged['amount_orig'].notna().all(), "Missing original values!"


# =========================
# 4. DIFFERENCE + DIRECTION
# =========================
merged['amount_diff'] = merged['amount_cf'] - merged['amount_orig']

merged['direction'] = np.where(
    merged['amount_diff'] > 0, 'increase',
    np.where(merged['amount_diff'] < 0, 'decrease', 'same')
)


# =========================
# 5. DIRECTION ANALYSIS
# =========================
summary = (
    merged
    .groupby(['predictive_model','prefix_length', 'fairness_opt', 'group_x', 'direction'])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

for col in ['increase', 'decrease', 'same']:
    if col not in summary.columns:
        summary[col] = 0

summary['total'] = summary[['increase', 'decrease', 'same']].sum(axis=1)

for col in ['increase', 'decrease', 'same']:
    summary[f'{col}_rate'] = summary[col] / summary['total']


# =========================
# 6. MEAN SHIFT (VERY IMPORTANT)
# =========================
mean_shift = (
    merged
    .groupby(['predictive_model', 'prefix_length', 'fairness_opt', 'group_x'])['amount_diff']
    .mean()
    .reset_index()
)

# =========================
# 7. FAIRNESS GAP
# =========================
gap_df = summary.pivot_table(
    index=['predictive_model', 'prefix_length', 'group_x'],  # ✅ added model
    columns='fairness_opt',
    values='increase_rate'
).reset_index()

gap_df.columns.name = None

gap_df = gap_df.rename(columns={
    True: 'increase_rate_fair',
    False: 'increase_rate_no_fair'
})

gap_df['gap'] = gap_df['increase_rate_fair'] - gap_df['increase_rate_no_fair']

# =========================
# 8. SAVE
# =========================
summary.to_csv("cf_direction_summary.csv", index=False)
mean_shift.to_csv("cf_mean_shift.csv", index=False)
gap_df.to_csv("cf_fairness_gap.csv", index=False)
merged.to_csv("cf_full_analysis.csv", index=False)


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

# --- FILTER ---
df = full_df[
    (full_df['dataset'] == 'bpi2012')
].copy()

# --- BUILD DATA ---
rows = []

for _, row in df.iterrows():
    changes = row["changes"]

    if isinstance(changes, str):
        changes = ast.literal_eval(changes)

    for attr, values in changes.items():
        orig = str(values.get("original"))
        cf = str(values.get("counterfactual"))

        rows.append({
            "model": row["model"],
            "changed_flag": orig != cf,
            "prefix_length": row["prefix_length"],
            "fairness_opt": row["fairness_opt"],
            "case": row["case"]
        })

attr_df = pd.DataFrame(rows)

# --- FILTER PROTECTED GROUP ---
attr_df_bpi = attr_df[
    attr_df['case'] == 'protected'
].copy()

# --- AGGREGATE ---
agg = (
    attr_df_bpi
    .groupby(["prefix_length", "model", "fairness_opt", "changed_flag"])
    .size()
    .reset_index(name="count")
)

# total per prefix_length + model + fairness
totals = agg.groupby(
    ["prefix_length", "model", "fairness_opt"]
)["count"].transform("sum")

agg["percentage"] = agg["count"] / totals * 100

# --- KEEP ONLY CHANGED ---
agg_changed = agg[agg["changed_flag"] == True]

# --- PREPARE PLOTS ---
fairness_vals = sorted(agg_changed["fairness_opt"].unique())
models = sorted(agg_changed["model"].unique())

n = len(fairness_vals)

fig, axes = plt.subplots(1, n, figsize=(6*n, 5), sharey=True)

# handle single subplot case
if n == 1:
    axes = [axes]

# --- PLOTTING ---
for i, fairness in enumerate(fairness_vals):
    ax = axes[i]

    subset = agg_changed[agg_changed["fairness_opt"] == fairness]

    for model in models:
        sub = subset[subset["model"] == model]

        ax.plot(
            sub["prefix_length"],
            sub["count"],
            marker='o',
            label=model
        )

    ax.set_title(f"Fairness: {fairness}")
    ax.set_xlabel("Prefix Length")
    if i == 0:
        ax.set_ylabel("Change Percentage (%)")

    ax.grid(True)

# --- LEGEND ---
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(
    handles,
    labels,
    title="Model",
    loc="upper center",
    ncol=len(models)
)

plt.tight_layout(rect=[0, 0, 1, 0.9])

plt.savefig(
    "figures/bpi2012_change_percentage_by_model_fairness.png",
    bbox_inches='tight'
)

plt.show()
```
